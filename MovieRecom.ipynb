```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
convert('[{"id": 18, "name": "Drama"}, {"id": 9648, "name": "Mystery"}, {"id": 27, "name": "Horror"}, {"id": 53, "name": "Thriller"}]')
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df.head(1)
```

```python
df['keywords'] = df['keywords'].apply(convert)
```

```python
df.head()
```

```python
def castConvert(obj):
    l=[]
    c = 0
    for i in ast.literal_eval(obj):
        if c < 3:
            l.append(i['name'])
            c+=1
        else:
            break
    return l
```

```python
df['cast'] = df['cast'].apply(castConvert)
```

```python
df.head(1)
```

```python
def dirConvert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
df['crew']=df['crew'].apply(dirConvert)
```

```python
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x : x.split())
```

```python
df
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['overview'] = df['overview'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['tag'] = df['genres']+df['overview']+df['keywords']+df['cast']+df['crew']
```

```python
df
```

```python
newdf = df[['movie_id','title','tag']]
```

```python
newdf['tag'] = newdf['tag'].apply(lambda x:" ".join(x))
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
    
```

```python
newdf['tag'][0]
```

```python
newdf['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
vector = cv.fit_transform(newdf['tag']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
cv = CountVectorizer(stop_words='english',max_features=5000)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
newdf[newdf['title'] == 'Batman Begins'].index[0]
```

```python
def recommend(movie):
    index = newdf[newdf['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(newdf.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```
